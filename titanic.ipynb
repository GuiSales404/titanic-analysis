```python
def std_method_plot(ds, feature):
  dt_mean, dt_std = np.mean(ds[f'{feature}']), np.std(ds[f'{feature}'])
  cut_off = dt_std * 3
  lower, upper = dt_mean - cut_off, dt_mean + cut_off
  idx = np.where((ds[f'{feature}'] < lower) | (ds[f'{feature}'] > upper))
  plt.subplots(figsize=(14,7))
  plt.subplot(1,2,1)
  sns.scatterplot(data = ds[f'{feature}'])
  sns.scatterplot(data = ds[f'{feature}'].iloc[idx],label='Outliers')
  plt.subplot(1,2,2)
  sns.scatterplot(ds[f'{feature}'],ds[f'{feature}'])
  sns.scatterplot(ds[f'{feature}'].iloc[idx],ds[f'{feature}'].iloc[idx],label='Outliers')
  plt.show();
```

```python
df['sobreviveu'] = df['sobreviveu'].replace([False, True], [0, 1])
df['genero'] = df['genero'].replace(['M', 'F'], [0, 1])
```

```python
from sklearn.preprocessing import Normalizer

X = df['idade'].fillna(df['idade'].mean()).to_numpy()
normalizer = Normalizer().fit([X])
res = normalizer.transform([X])
df['idade'] = res.reshape(-1)
```

```python
X = df['preco_bilhete'].to_numpy()
normalizer = Normalizer().fit([X])
result = normalizer.transform([X])
df['preco_bilhete'] = result.reshape(-1)
```

```python
X = df['irmaos_conjuges_abordo'].to_numpy()
normalizer = Normalizer().fit([X])
result = normalizer.transform([X])
df['irmaos_conjuges_abordo'] = result.reshape(-1)
```

```python
X = df['pais_filhos_abordo'].to_numpy()
normalizer = Normalizer().fit([X])
result = normalizer.transform([X])
df['pais_filhos_abordo'] = result.reshape(-1)
```

```python
X = df.drop(axis=1, columns=['sobreviveu', 'classe_bilhete', 'numero_bilhete', 'cabine_numero', 'local_embarque']).reset_index()
y = df['sobreviveu']
```

```python
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import cross_validate
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
models = {}

# Logistic Regression
from sklearn.linear_model import LogisticRegression
models['Logistic Regression'] = LogisticRegression(max_iter=1000)

# Decision Trees
from sklearn.tree import DecisionTreeClassifier
models['Decision Trees'] = DecisionTreeClassifier()

# Random Forest
from sklearn.ensemble import RandomForestClassifier
models['Random Forest'] = RandomForestClassifier()

# Naive Bayes
from sklearn.naive_bayes import GaussianNB
models['Naive Bayes'] = GaussianNB()
```

```python
accuracy, precision, recall = {}, {}, {}
y_pred = []

for key in models.keys():
    
    # Fit the classifier
    models[key].fit(X_train, y_train)
    
    # Make predictions
    predictions = models[key].predict(X_test)
    y_pred.append(predictions)

    # Calculate metrics
    accuracy[key] = accuracy_score(predictions, y_test)
    precision[key] = precision_score(predictions, y_test)
    recall[key] = recall_score(predictions, y_test)
```

```python
df_model = pd.DataFrame(index=models.keys(), columns=['Accuracy', 'Precision', 'Recall'])
df_model['Accuracy'] = accuracy.values()
df_model['Precision'] = precision.values()
df_model['Recall'] = recall.values()

df_model
```

```python
ax = df_model.plot.barh()
ax.legend(
    ncol=len(models.keys()), 
    bbox_to_anchor=(0, 1), 
    loc='lower left', 
    prop={'size': 14}
)
plt.tight_layout()
```

```python
def conf_mat(model_number):
    models = ['Logistic Regression', 'Decision Trees', 'Random Forest', 'Naive Bayes']
    print(models[model_number])
    cm = confusion_matrix(y_test, y_pred[model_number])
    ax= plt.subplot()
    sns.heatmap(cm, annot=True, fmt='g', ax=ax); 
    ax.set_xlabel('Predicted labels');ax.set_ylabel('True labels'); 
    ax.set_title('Confusion Matrix'); 
    ax.xaxis.set_ticklabels(['sobreviveu', 'não sobreviveu']); ax.yaxis.set_ticklabels(['sobreviveu', 'não sobreviveu']);
```
